```python
class sales(Employee):
  def __init__(self,name,emp_id,salary,gross_sales,commission_rate):
    super().__init__(name,emp_id,salary)
    self._gross_sales = gross_sales
    self._commission_rate = commission_rate

    def get_total_salary(self):
      return self.salary+ self._gross_sales*self._commission_rate

    def printt(self):
      super().printt()
      print('gross sales',self._gross_sales,'commission rate',self._commission_rate)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cProfile
import timeit
import time
```

```python
class DataAnalyzer:
  def __init__(self,num_samples):
    self.df = None
    self.num_samples = num_samples
    self.data = None
  def generate_data(self):
    self.data = np.random.randint(1,100,size = self.num_samples)
    self.df = pd.DataFrame(self.data,columns = ['values'])
    print(f'data generated{self.df}')
  def analyze_data(self):
    mean = self.df['values'].mean()
    median = self.df['values'].median()
    std_dev = self.df['values'].std()
    print(f'mean :{mean},median: {median},standard deviation {std_dev}')
```

```python
def main():
  num_samples = 500
  analyser = DataAnalyzer(num_samples)
  print(analyser.generate_data())
  print(analyser.analyze_data())

if __name__ == '__main__':
  main()
  excution_time = timeit.timeit(main,number = 1)
  print('time taken',excution_time)
```
